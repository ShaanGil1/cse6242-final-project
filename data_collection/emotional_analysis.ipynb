```python
import pandas as pd
import numpy as np
path = "D:/CS6242/covid19_twitter/full_data_withSentimentandStates.csv"
df = pd.read_csv(path)
```

```python
df["Happy"] = None
df["Angry"] = None
df["Surprise"] = None
df["Sad"] = None
df["Fear"] = None
interval = 50000
start = 450000
end = start + interval
df = df[start:end]
print(len(df))
```

```python
text = "I hate everything"
te.get_emotion(text)
```

```python
new_col = []
count = 0
for index, row in df.iterrows():
    count+=1
    if index % 100 == 0:
        print("Progress: " + str(index/len(df)))
    if row["text"] == "error":
        new_col.append("error")
    else:
        emotion = te.get_emotion(row["text"])
        new_col.append(emotion)
```

```python
print(new_col)
for i in range(0, len(new_col)):
    df.loc[i + start, "Happy"] = new_col[i]["Happy"]
    df.loc[i + start, "Angry"] = new_col[i]["Angry"]
    df.loc[i + start, "Surprise"] = new_col[i]["Surprise"]
    df.loc[i + start, "Sad"] = new_col[i]["Sad"]
    df.loc[i + start, "Fear"] = new_col[i]["Fear"]
print(df)
    
```

```python
out = pd.DataFrame.to_csv(df, path_or_buf="D:\CS6242\covid19_twitter\\batch_" + str(start) + "-" + 
                          str(end) + ".csv")
```
