```python
import tweepy
import pandas as pd
import os
import math
```

```python
path = "D:\CS6242\covid19_twitter\dailies"
files = []
for filename in os.listdir(path):
    f = os.path.join(path, filename)
    for filen in os.listdir(f):
        g = os.path.join(f, filen)
        if g[-17:] == "clean-dataset.tsv":
            files.append((f[-10:], g))
print(files)
```

```python
num_row = 0
limit = 0

batches = []
num_rows = df.shape[0]
limit = math.ceil(num_rows/100)
for i in range(limit - 1):
    batches.append(df.iloc[i*100:(i+1)*100])
batches.append(df.iloc[(limit - 1)*100:])

print(batches[0])
```

```python
print(limit)
```

```python
final_df = []
count = 0
for batch in batches:
    batch = batch['tweet_id'].tolist()
    tweets_info = client.get_tweets(ids=batch, 
                                tweet_fields=["created_at", "public_metrics", "author_id", "lang", "source", "possibly_sensitive"],
                               expansions=["geo.place_id"],
                               place_fields=["geo", "full_name", "id", "name", "place_type", "contained_within"])
    box = []
    loc_fname = []
    loc_type = []
    loc_name = []
    loc_id = []
    for tweet in tweets_info.includes["places"]:
        box.append(tweet["geo"]['bbox'])
        loc_fname.append(tweet["full_name"])
        loc_type.append(tweet["place_type"])
        loc_name.append(tweet["name"])
        loc_id.append(tweet["id"])
    
    geo_id = []
    tweet_id = []
    text = []
    created_at = []
    retweet = []
    reply = []
    likes = []
    quote = []
    author_id = []
    source = []
    lang = []
    sensitive = []
    for tweet in tweets_info.data:
        metrics = tweet["public_metrics"]
        if (tweet.geo == None):
            continue
        geo_id.append(tweet.geo["place_id"])
        tweet_id.append(tweet["id"])
        text.append(tweet["text"])
        created_at.append(tweet["created_at"])
        retweet.append(metrics["retweet_count"])
        reply.append(metrics['reply_count'])
        likes.append(metrics['like_count'])
        quote.append(metrics['quote_count'])
        author_id.append(tweet['author_id'])
        source.append(tweet["source"])
        lang.append(tweet["lang"])
        sensitive.append(tweet["possibly_sensitive"])
    pcount = 0
    for i in range(0, len(geo_id)):
        for j in range(0, len(loc_id)):
            if loc_id[j] == geo_id[i]:
                pcount+=1
                #print("I counter: " + str(i))
                #print("P counter: " + str(pcount))
                final_df.append((tweet_id[i], text[i], loc_id[j], created_at[i], retweet[i], reply[i], likes[i], quote[i], author_id[i],
                                     source[i], lang[i], sensitive[i], box[j], loc_name[j], loc_type[j], loc_fname[j]))
        
    for error in tweets_info[2]:
        final_df.append((error["value"], 'error', 'error', 'error', 'error', 'error', 'error', 'error',
                           'error', 'error', 'error', 'error', 'error', 'error', 'error', 'error'))
    count+=1
    if count % 5 == 0:
        print("Progress : %2.4f " % (count/limit * 100))
    #if count==5:
    #    break
```

```python
finaldf = pd.DataFrame(final_df)
```

```python
finaldf.columns = ["tweet_id", "text", "loc_id", "created_at", "retweet_count", "reply_count",
                  "likes_count", "quote_count", "author_id", "source", "lang", "is_sensitive", "bbox",
                  "loc_name", "loc_type", "loc_fname"]
```
