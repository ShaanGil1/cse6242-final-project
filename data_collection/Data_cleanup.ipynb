```python
#v6 start
#calculating pearson coefficient for sentiment and emotion for each state
path = os.path.join(directory, 'weekly_vax_sentiment_emotion.csv')
df = pd.read_csv(path)
axis1 = ['neg', 'neu', 'pos', 'compound', 'Happy', 'Angry', 'Surprise', 'Sad', 'Fear']
df.rename(columns={'Administered_Dose1_Pop_Pct':'Dose1', 'Series_Complete_Pop_Pct':'Dose2', 'Additional_Doses_Vax_Pct':'Dose3'}, inplace=True)
print(df)
axis2 = ['Dose1', 'Dose2', 'Dose3']
```

```python
state_averages_df = df.groupby(['state']).mean().reset_index()
test = df.iloc[5550]
print((test['Angry'] - state_averages_df[state_averages_df['state'] == test['state']]['Angry']).iloc[0])
test['test'] = (test['Angry'] - state_averages_df[state_averages_df['state'] == test['state']]['Angry'].iloc[0])
print(test)
for a1 in axis1:
    df['avg'+a1] = df.apply(lambda x: (state_averages_df[state_averages_df['state'] == x['state']][a1]).iloc[0], axis=1)
for a2 in axis2:
    df['avg'+a2] = df.apply(lambda x: (state_averages_df[state_averages_df['state'] == x['state']][a2]).iloc[0], axis=1)
print(df)
```

```python
for a1 in axis1:
    df['x'+a1] = df.apply(lambda x: x[a1] - x['avg'+a1], axis=1)
for a2 in axis2:
    df['y'+a2] = df.apply(lambda x: x[a2] - x['avg'+a2], axis=1)
for a1 in axis1:
    df['x2'+a1] = df.apply(lambda x: x['x'+a1]**2, axis=1)
for a2 in axis2:
    df['y2'+a2] = df.apply(lambda x: x['y'+a2]**2, axis=1)
for a1 in axis1:
    for a2 in axis2:
        df['xy'+a1+a2] = df.apply(lambda x: x['x'+a1] * x['y'+a2], axis=1)
# print(df)
```

```python
r_df = df.groupby('state').sum()
rs = []
for a1 in axis1:
    for a2 in axis2:
        rs.append('r'+a1+a2)
        r_df['r'+a1+a2] = r_df.apply(lambda x: x['xy'+a1+a2]/np.sqrt(x['x2'+a1]*x['y2'+a2]), axis=1)
r_df = r_df[rs]
```

```python
r_path = os.path.join(directory, 'r_by_state.csv')
r_df.to_csv(r_path, index=False)
#v6 end
```
